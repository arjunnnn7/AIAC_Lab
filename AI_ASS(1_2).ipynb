```python
#Task 1: Factorial Without Functions (No Modularization)
# Factorial without using any user-defined function
# Logic written directly in main execution flow

n = int(input("Enter a number: "))

factorial = 1

for i in range(1, n + 1):
    factorial *= i

print("Factorial of", n, "is:", factorial)
```

```python
#Task 2: Optimized & Cleaned Version (Still No Functions)
# Optimized factorial logic without functions
# Reduced variables and improved readability

n = int(input("Enter a number: "))

result = 1
for i in range(2, n + 1):
    result *= i

print(f"Factorial of {n} is: {result}")
```

```python
#Task 3: Modular Design – Factorial Using Function
# Factorial using a user-defined function
# Modular and reusable design

def calculate_factorial(number):
    result = 1
    for i in range(1, number + 1):
        result *= i
    return result


n = int(input("Enter a number: "))
output = calculate_factorial(n)
print("Factorial of", n, "is:", output)
```

Task 4: Comparative Analysis –

Procedural vs Modular AI Code Procedural code without functions is useful for small, quick scripts, but it lacks flexibility and clarity. Function-based modular code improves readability, reusability, and maintainability. For large projects and collaborative development, modular design is always preferred. AI-generated code becomes more reliable and easier to evaluate when it follows a modular approach.

```python
# Task 5: Iterative vs Recursive Implementations
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


num = int(input("Enter a number: "))
print("Iterative Factorial:", factorial_iterative(num))
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)


num = int(input("Enter a number: "))
print("Recursive Factorial:", factorial_recursive(num))
```
