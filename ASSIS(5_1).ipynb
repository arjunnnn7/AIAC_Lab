```python

#Task 1: Privacy in API Usage
import os
import requests

# API key is stored as an environment variable
api_key = os.getenv("WEATHER_API_KEY")

if api_key is None:
    print("API key not found. Please set WEATHER_API_KEY.")
else:
    city = input("Enter city name: ")
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"

    response = requests.get(url)
    data = response.json()

    if response.status_code == 200:
        print("Temperature:", data["main"]["temp"])
    else:
        print("Error fetching weather data")
```

```python
#Task 2: Privacy & Security in File Handling
import hashlib

name = input("Enter name: ")
email = input("Enter email: ")
password = input("Enter password: ")

# Hashing password
hashed_password = hashlib.sha256(password.encode()).hexdigest()

with open("users_secure.txt", "a") as file:
    file.write(f"{name},{email},{hashed_password}\n")

print("User data stored securely.")
```

```python
#Task 3: Transparency in Algorithm Design
def is_armstrong(number):
    """
    This function checks whether a number is an Armstrong number.
    """
    digits = len(str(number))   # Count number of digits
    temp = number
    total = 0

    # Calculate sum of digits raised to power of number of digits
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == number


n = int(input("Enter a number: "))
if is_armstrong(n):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
#Task 4: Transparency in Algorithm Comparison
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]
    left = [x for x in arr[1:] if x <= pivot]
    right = [x for x in arr[1:] if x > pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)
arr = [5, 3, 8, 4, 2]
print("Bubble Sort:", bubble_sort(arr.copy()))
print("Quick Sort:", quick_sort(arr.copy()))
```

```python
#Task 5: Transparency in AI Recommendations
def recommend_product(user_interest):
    products = {
        "laptop": "Recommended because it suits programming and productivity tasks.",
        "mobile": "Recommended for communication and daily usage.",
        "headphones": "Recommended for online classes and media consumption."
    }

    if user_interest in products:
        print("Product:", user_interest)
        print("Reason:", products[user_interest])
    else:
        print("No recommendation available for this interest.")


interest = input("Enter your interest (laptop/mobile/headphones): ")
recommend_product(interest)
```
